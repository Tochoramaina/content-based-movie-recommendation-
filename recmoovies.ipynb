genres has entries made up of dictionaries composed of keys and values. we will remove keys and remain with values which are vital in our analysis

```python
# evaluating texts before we cleanse them
from ast import literal_eval
literal_eval('{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name":"Science Fiction"}')
```

below is a function that will remove keys that is 'id' and 'name' and leave us with values like action, adventure, fantasy and science fiction

```python
# a function to cleanse our data
def convert(obj):
    L = []
    for i in literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
#applying our function to genre column
df['genres'] = df['genres'].apply(convert)
```

```python
# applying the same fuction to these columns since they have enties that have the same problem
df['production_companies'] = df['production_companies'].apply(convert)
df['spoken_languages'] = df['spoken_languages'].apply(convert)
df['keywords'] = df['keywords'].apply(convert)
df['cast'] = df['cast'].apply(convert)
df['crew'] = df['crew'].apply(convert)
```

```python
df3['overview'] = df3['overview'].apply(lambda x : x if isinstance(x, list) else [x])
```

```python
# combining genres,overview,keywords, cast into one column; tags
df3['tags'] = df3['genres'] + df3['overview'] + df3['keywords'] + df3['cast']
# a new dataframe made up of 3 columns
new_df = df3[['movie_id', 'title', 'tags']]
new_df.head()
```

```python
# converting the datatype of tags into strings
new_df['tags'] = new_df['tags'].astype(str)
```

```python
# changing tags from proper case to lower case
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

Bagging of words

Countvectorizer will convert text to a sparse matrix of numrical data. stop_words will remove words that don't add meaning to our text like articles,prepositions and interjections from our text.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
# applying count vectorizer to our text
count_vectors = cv.fit_transform(new_df['tags']).toarray()
```

we will modify words to their basic format example walking => walk, eaten => eat using WordNetLemmatizer

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
# a function modify words
def lemmatizing(text):
    sentence = []
    for i in text.split():
        sentence.append(lemmatizer.lemmatize(i))
    return " ".join(sentence)
```

```python
# applying our function to tags column
new_df['tags'].apply(lemmatizing)
```

```python

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(count_vectors)
```

```python
similarity
```

```python
# creating indices of  movies and shows
indices = pd.Series(new_df.index, index=new_df['title'])
indices
```

Movie Recommendations

below is a function that we will use to make recommendations on a movie or show that a user prefers based on the content that is in the movie.this will depend on tags column.

```python
# a function that takes in movie title as inputs and simillar movies as output
def get_recommedations(title):
    idx = indices[title]
    sim_scores_desc = list(enumerate(similarity[idx]))
    sim_scores_desc = sorted(sim_scores_desc, key=lambda x : x[1], reverse=True)
    sim_scores_desc = sim_scores_desc[1:11]
    movie_indices = [i[0] for i in sim_scores_desc]
    return new_df[['movie_id', 'title', 'tags']].iloc[movie_indices]
```

```python
# movie recommedations similar to spectre
get_recommedations("Spectre")
```

we can see that all movies recommended are simillar to spectre in relation to genre, cast, overview and keywords
